```python
import pandas as pd
import numpy as np
import ast
import re

df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
# source project id
# name
# country raw
# funder name?
df["unique_id"] = range(1, len(df) + 1)
```

```python
# Drop rows where 'column_name' is null
df_1 = df.dropna(subset=["source_project_id"])
df_new = df_1.dropna(subset=["name"])
df_new["recipient_name"] = df_new["recipient_name"].fillna("UNKNOWN")
# Save cleaned file (optional)
df_new.to_csv("data_cleaned.csv", index=False)

print("Nulls removed successfully.")
```

```python
# Check if composite key is unique
is_unique = not df.duplicated(subset=["source_project_id", "name", "recipient_name"]).any()

print("Composite is unique:", is_unique)
```

```python
# Count the number of null values for each column
def null_count_table(df: pd.DataFrame) -> pd.DataFrame:
    result = (
        df.isna()
          .sum()
          .reset_index()
          .rename(columns={"index": "Column", 0: "Null_Count"})
    )
    
    return result
```

```python
print(null_count_table(df_new))
```
