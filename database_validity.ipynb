## Check for Null columns

```python
null_columns = df.columns[df.isna().any()]
print(null_columns.tolist())
```

```python
# Count the number of null values for each column
def null_count_table(df: pd.DataFrame) -> pd.DataFrame:
    result = (
        df.isna()
          .sum()
          .reset_index()
          .rename(columns={"index": "Column", 0: "Null_Count"})
    )
    
    return result
```

```python
print(null_count_table(df))
```

## Primary Key Testing

```python
# Check whether a set of columns is unique
# Input: df: Dataframe, columns: list of column names, ['col1', 'col2', ...]
def is_valid_primary_key(df: pd.DataFrame, columns) -> bool:
    if isinstance(columns, str):
        columns = [columns]
    return not df.duplicated(subset=columns).any()
```

```python
# Fill in null values with placeholder value
# Add autoincrement 'id' column
new_df = df.fillna("Nan")
new_df['id'] = range(1, len(new_df) + 1)
```

```python
# Change primary_key for testing
primary_key = ['source_project_id', 'name']
if is_valid_primary_key(new_df, primary_key):
    print("[" + ', '.join(primary_key) + "] \033[32mis\033[32m a valid primary key")
else: print("[" + ', '.join(primary_key) + "] \033[31mis not\033[0m a valid primary key")
```
